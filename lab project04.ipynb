Generate two sinusoids of length 10 and fd = 0.1 with variable phase. 2. Apply correlation and check for certain
properties such as magnitude and location of maximum correlation with varying phases.

```python
import numpy as np
import matplotlib.pyplot as plt
# Define parameters
n = np.arange(10) # n = [0:9]
ph1 = 0 # Phase shift for x
ph2 = 0 # Phase shift for y
# Generate signals
x = np.sin(2 * np.pi * 0.1 * n + ph1)
org_x = 1
nx = np.arange(len(x)) - org_x + 1 # Adjust time index for x
y = np.sin(2 * np.pi * 0.1 * n + ph2)
org_y = 1
ny = np.arange(len(y)) - org_y + 1 # Adjust time index for y
# Compute cross-correlation
rxy = np.correlate(x, y, mode='full')
nr = np.arange(nx[0] - ny[-1], nx[-1] - ny[0] + 1)
# Find maximum correlation and its lag
maxR = np.max(rxy)
indR = np.argmax(rxy)
lag_at_max_corr = nr[indR]
# Display results
print(f'The correlation at lag zero is: {rxy[np.where(nr == 0)][0]}.')
print(f'The maximum correlation is at lag {lag_at_max_corr}.')
# Plot signals and cross-correlation
plt.figure(figsize=(8, 10))
# Plot x(n)
plt.subplot(3, 1, 1)
plt.stem(nx, x)
plt.xlabel('Time index n')
plt.ylabel('Amplitude')
plt.xlim([nx[0] - 1, nx[-1] + 1])
plt.title('Signal x(n)')
plt.grid()
# Plot y(n)
plt.subplot(3, 1, 2)
plt.stem(ny, y)
plt.xlabel('Time index n')
plt.ylabel('Amplitude')
plt.xlim([ny[0] - 1, ny[-1] + 1])
plt.title('Signal y(n)')
plt.grid()
# Plot cross-correlation
plt.subplot(3, 1, 3)
plt.stem(nr, rxy)
plt.xlabel('Time index n')
plt.ylabel('Amplitude')
plt.xlim([nr[0] - 1, nr[-1] + 1])
plt.title('Cross Correlation')
plt.grid()
plt.tight_layout()
plt.show()
```

# **Task -1**

Now modify the phase of the second signal to pi/2 (it will make it cosine) and observe the correlation at lag
zero. Modify the phase again to pi and observe.
1. Check for auto-correlation (ph1 = ph2) that the lag zero value gives the energy of the Signal.
2. Observe that the commutative property does not hold.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

n = np.arange(10)
ph1 = 0
ph2_list = [np.pi/2, np.pi]

for ph2 in ph2_list:
    x = np.sin(2 * np.pi * 0.1 * n + ph1)
    y = np.sin(2 * np.pi * 0.1 * n + ph2)

    rxy = np.correlate(x, y, mode='full')
    nr = np.arange(-len(n) + 1, len(n))

    maxR = np.max(rxy)
    indR = np.argmax(rxy)
    lag_at_max_corr = nr[indR]

    rxx = np.correlate(x, x, mode='full')
    energy_x = rxx[len(n) - 1]

    ryx = np.correlate(y, x, mode='full')

    print(f'\nFor phase shift ph2 = {ph2} ({"cosine" if ph2 == np.pi/2 else ""})')
    print(f'The correlation at lag zero is: {rxy[np.where(nr == 0)][0]}.')
    print(f'The maximum correlation is at lag {lag_at_max_corr}.')
    print(f'Auto-correlation at lag zero: {energy_x}.')
    print(f'Cross-correlation does not satisfy commutativity: {not np.allclose(rxy, ryx[::-1])}')

    plt.figure(figsize=(8, 10))

    plt.subplot(3, 1, 1)
    plt.stem(n, x)
    plt.xlabel('Time index n')
    plt.ylabel('Amplitude')
    plt.title('Signal x(n)')
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.stem(n, y)
    plt.xlabel('Time index n')
    plt.ylabel('Amplitude')
    plt.title(f'Signal y(n) with phase shift {ph2}')
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.stem(nr, rxy)
    plt.xlabel('Lag index')
    plt.ylabel('Amplitude')
    plt.title('Cross-Correlation x[n] * y[n]')
    plt.grid()

    plt.tight_layout()
    plt.show()
```
