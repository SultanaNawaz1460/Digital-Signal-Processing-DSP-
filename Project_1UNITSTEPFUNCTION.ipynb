# UNIT STEP FUNCTION

```python
import numpy as np
import matplotlib.pyplot as plt

# Time array for n = 3 to n = 7
n = np.arange(3, 8)

# Unit step function (u(n) = 1 for all n >= 0)
u = np.ones_like(n)

# Plotting the unit step function
plt.figure(figsize=(10, 6))
plt.stem(n, u, basefmt=" ")
plt.title("Discrete Time Unit Step Function (n=3 to n=7)")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
